```
## Path Defintions
path_videos = F"/content/gdrive/My Drive/deep_learning/Project/Videos"
file_name = 'IMG_3561.MOV'
path_images = f"/content/gdrive/My Drive/deep_learning/Project/IMAGES" 
```

```
## Imports 
import numpy as np
import torch
from torch import nn
import torch.nn.functional as F
import os
import matplotlib.pyplot as plt
import pandas as pd
import os
import cv2
from PIL import Image
```

```
## Videos in path:
os.listdir(path_videos)
```

```
## Convert video to images
print(file_name)    
cap = cv2.VideoCapture(path_videos+'/'+file_name)

# Check if camera opened successfully
if (cap.isOpened()== False): 
    print("Error opening video stream or file")

# Read until video is completed
frame_number = 0
while(cap.isOpened()):
      # Capture frame-by-frame
      ret, Newframe = cap.read()
      if Newframe is None:
          break
      frame_number+=1
      Image.fromarray(cv2.cvtColor(Newframe, cv2.COLOR_BGR2RGB)).save("{}/{}_{}.jpeg".format(path_images, file_name[:-4], frame_number))
      if frame_number%100==0:
          print(frame_number)
          plt.imshow(cv2.cvtColor(Newframe, cv2.COLOR_BGR2RGB),cmap='gray')
          plt.show()
            
# When everything done, release the video capture object
cap.release()
```
